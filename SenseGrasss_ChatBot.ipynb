```python
import re

def get_crop_info(crop):
    crop_info = {
        'wheat': "Wheat requires a cool climate with temperatures ranging between 10°C to 24°C. It needs well-drained loamy soil with a pH of 6.5 to 7.5. It requires moderate rainfall, and the ideal sowing time is in the winter season.",
        'rice': "Rice needs a warm and humid climate with temperatures ranging between 20°C to 37°C. It thrives in flooded fields and requires heavy rainfall or irrigation. It grows well in clayey or loamy soil with good water retention capacity.",
        'jowar': "Jowar, also known as sorghum, is a hardy crop that can grow in semi-arid to arid regions. It requires temperatures between 25°C to 32°C and grows well in well-drained sandy or loamy soil. Jowar is drought-tolerant and requires minimal water."
    }
    return crop_info.get(crop.lower(), "Sorry, I don't have information about that crop.")

def detect_crop_in_sentence(sentence):
    crops = re.findall(r'\b(wheat|rice|jowar)\b', sentence, flags=re.IGNORECASE)
    if crops:
        return crops[0].lower()
    else:
        return None

def main():
    print("Welcome to the Agriculture Chatbot!")
    print("Ask any question related to agriculture. Type 'quit' to exit.")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("Chatbot: Goodbye!")
            break
        else:
            crop = detect_crop_in_sentence(user_input)
            if crop:
                response = get_crop_info(crop)
                print("Chatbot:", response)
            else:
                print("Chatbot: Please specify a crop (wheat, rice, or jowar) in your question.")

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

# Crop images for each crop
wheat_image = plt.imread("/content/wheat_sg.jpeg")
rice_image = plt.imread("/content/Rice_sg.jpeg")
jowar_image = plt.imread("/content/jowar_sg.jpeg")

# Dictionary to map crop names to their images
crop_images = {
    'wheat': wheat_image,
    'rice': rice_image,
    'jowar': jowar_image
}

crop_info = {
    'wheat': {
        'growing_conditions': "Wheat requires a cool climate with temperatures ranging between 10°C to 24°C. It needs well-drained loamy soil with a pH of 6.5 to 7.5. It requires moderate rainfall, and the ideal sowing time is in the winter season.",
        'irrigation': "Wheat generally requires irrigation, especially during the flowering and grain-filling stages. However, excessive water can lead to lodging and disease problems.",
        'harvesting': "Wheat is typically harvested when the grains are fully ripe and dry. This usually occurs in late spring or early summer, depending on the region."
    },
    'rice': {
        'growing_conditions': "Rice needs a warm and humid climate with temperatures ranging between 20°C to 37°C. It thrives in flooded fields and requires heavy rainfall or irrigation. It grows well in clayey or loamy soil with good water retention capacity.",
        'irrigation': "Rice cultivation often involves flooding or constant standing water in the fields. Irrigation is necessary, especially during the dry season or in areas with insufficient rainfall.",
        'harvesting': "Rice is typically harvested when the grains have reached full maturity and the moisture content is around 20%. This usually occurs between 3 to 6 months after planting, depending on the variety and growing conditions."
    },
    'jowar': {
        'growing_conditions': "Jowar, also known as sorghum, is a hardy crop that can grow in semi-arid to arid regions. It requires temperatures between 25°C to 32°C and grows well in well-drained sandy or loamy soil. Jowar is drought-tolerant and requires minimal water.",
        'irrigation': "Jowar cultivation may require irrigation during dry periods, but it is relatively drought-tolerant compared to other crops. It is important to avoid overwatering, as it can lead to root rot.",
        'harvesting': "Jowar is typically harvested when the grains are fully ripe and dry. This usually occurs around 90 to 120 days after planting, depending on the variety and growing conditions."
    }
}

def get_crop_info(crop, aspect):
    if crop.lower() in crop_info:
        if aspect:
            # If a specific aspect is mentioned, return information about that aspect
            if aspect.lower() in crop_info[crop.lower()]:
                return crop_info[crop.lower()][aspect.lower()]
            else:
                return "Sorry, I don't have information about that aspect."
        else:
            # If no specific aspect is mentioned, return information about all aspects
            return "\n".join([f"{key.capitalize()}: {value}" for key, value in crop_info[crop.lower()].items()])
    else:
        return "Sorry, I don't have information about that crop."


def main():
    print("Welcome to the SenseGrass Chatbot, I am Groot!")
    print("Ask any question related to agriculture. Type 'Bye Groot' to exit.")

    # Display the welcome image
    welcome_image = plt.imread("/content/welcome_sg.png")
    plt.imshow(welcome_image)
    plt.axis('off')
    plt.show()

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'bye groot':
            print("Chatbot: Goodbye!")
            break
        else:
            crop, aspect = None, None
            for key in crop_info.keys():
                if key in user_input.lower():
                    crop = key
                    break
            if not crop:
                print("Chatbot: Please specify a crop (wheat, rice, or jowar) in your question.")
                continue
            else:
                # Check for aspect words in the user input
                aspect_words = ['growing', 'irrigation', 'harvesting']
                for word in aspect_words:
                    if word in user_input.lower():
                        aspect = word
                        break
            output = get_crop_info(crop, aspect)
            plt.imshow(crop_images[crop])
            plt.axis('off')
            plt.show()
            print("Chatbot:", output)

if __name__ == "__main__":
    main()
```
